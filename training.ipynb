## individuo le variabili e le trasformo in label

```python
s=(heart_failure_clinical_records.dtypes == 'object')
object_cols = list(s[s].index)

print (object_cols)
```

```python
label_heart_failure_clinical_records=heart_failure_clinical_records.copy()
label_encoder=LabelEncoder()
for col in object_cols:
    label_heart_failure_clinical_records[col]=label_encoder.fit_transform(label_heart_failure_clinical_records[col])
```

```python
label_heart_failure_clinical_records
```

```python
y=label_heart_failure_clinical_records.DEATH_EVENT
X=label_heart_failure_clinical_records.drop("DEATH_EVENT", axis=1)
```

```python
y
```

```python
X
```

```python
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=0)
```

```python
train_X
```

```python
train_y
```

## modellazione e allenamento

```python
#Definisco il modello 
heart_failure_clinical_records_model=DecisionTreeRegressor(random_state=1)
#Alleno il modello
heart_failure_clinical_records_model.fit(train_X,train_y)
```

```python
#validazione e calcolo del Mean absosulte error
val_pred=heart_failure_clinical_records_model.predict(val_X)

print(mean_absolute_error(val_y,val_pred))
```

## Accuracy

```python
acc_heart_failure_clinical_records_model=accuracy_score(val_y,val_pred)
acc_heart_failure_clinical_records_model
```
